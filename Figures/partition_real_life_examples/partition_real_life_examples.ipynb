```r
# check if the value of a,b,c is appropriate
options(repr.plot.width=10, repr.plot.height=8)

clustering_file <- '../../partition_real_life_examples/PBMC_SnapATAC_ndim15_seed0_r0.15.tsv'
ground_truth_file <- '../../partition_real_life_examples/PBMC_ground_truth.tsv'

df_label <- read.table(clustering_file, header = T, sep="\t", comment.char = "")
df_ground_truth <- read.table(ground_truth_file, header = T, sep="\t", comment.char = "")

df <- merge(df_label, df_ground_truth, by="barcode")
ground_truth <- as.factor(df$ground_truth)
clusterings <- as.factor(df$clusterings)

p1 <- cross_table_plot(ground_truth, clusterings, a=2, b=3.8, c=3.1, m=0, n=0.35)
p1
```

```r
# check if the value of a,b,c is appropriate
options(repr.plot.width=10, repr.plot.height=8)

ground_truth_file <- '../../partition_real_life_examples/PBMC_ground_truth.tsv'
clustering_file <- '../../partition_real_life_examples/PBMC_SnapATAC_ndim15_seed0_r0.2.tsv'

df_label <- read.table(clustering_file, header = T, sep="\t", comment.char = "")
df_ground_truth <- read.table(ground_truth_file, header = T, sep="\t", comment.char = "")

df <- merge(df_label, df_ground_truth, by="barcode")
ground_truth <- as.factor(df$ground_truth)
clusterings <- as.factor(df$clusterings)

p2 <- cross_table_plot(ground_truth, clusterings, a=2, b=3.8, c=3.1, m=0, n=0.35)
p2
```

```r
options(repr.plot.width=20, repr.plot.height=8)
combined <- (p1 + p2) +
  plot_annotation(tag_levels = 'a') &
  theme(plot.tag = element_text(size = 30, hjust = -2, vjust = 1.5))

print(combined)
```
