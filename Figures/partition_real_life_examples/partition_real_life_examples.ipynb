```r
# check if the value of a,b,c is appropriate
options(repr.plot.width=10, repr.plot.height=8)

clustering_file <- '../../data/partition_real_life_examples/PBMC_SnapATAC_ndim15_seed0_r0.15.tsv'
ground_truth_file <- '../../data/partition_real_life_examples/PBMC_ground_truth.tsv'

df_label <- read.table(clustering_file, header = T, sep="\t", comment.char = "")
df_ground_truth <- read.table(ground_truth_file, header = T, sep="\t", comment.char = "")

df <- merge(df_label, df_ground_truth, by="barcode")
ground_truth <- as.factor(df$ground_truth)
clusterings <- as.factor(df$clusterings)

p1 <- cross_table_plot(ground_truth, clusterings, a=2, b=3.8, c=3.1, m=0, n=0.35)
p1 <- annotate_figure(p1, top = text_grob("15 classes, 10 clusters", face = "bold", size = 22))
p1
```

```r
# check if the value of a,b,c is appropriate
options(repr.plot.width=10, repr.plot.height=8)

clustering_file <- '../../data/partition_real_life_examples/PBMC_SnapATAC_ndim15_r0.45.tsv'
ground_truth_file <- '../../data/partition_real_life_examples/PBMC_ground_truth.tsv'

df_label <- read.table(clustering_file, header = T, sep="\t", comment.char = "")
df_ground_truth <- read.table(ground_truth_file, header = T, sep="\t", comment.char = "")

df <- merge(df_label, df_ground_truth, by="barcode")
ground_truth <- as.factor(df$ground_truth)
clusterings <- as.factor(df$clusterings)

p3 <- cross_table_plot(ground_truth, clusterings, a=2, b=3.8, c=3.1, m=0, n=0.35)
p3 <- annotate_figure(p3, top = text_grob("15 classes, 15 clusters", face = "bold", size = 22))
p3
```
